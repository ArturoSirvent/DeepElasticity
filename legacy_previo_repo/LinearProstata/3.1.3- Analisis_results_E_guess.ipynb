Simplementa para analizar los resultado de la predicci√≥n de E anterior.   

```python
import pandas as pd 
import json 
import matplotlib.pyplot as plt  
import seaborn as sns 
import numpy as np 
```

```python
RESULTS_DIR="results/3.1.2/2000_logs.json"
```

```python
with open(RESULTS_DIR,"r") as fil:  
    raw_results=json.load(fil)

    
```

```python
plt.figure(figsize=(13,7))
for j,E in enumerate(raw_results.keys()):
    plt.subplot(2,2,j+1)
    for i in raw_results[E].keys():  
        plt.plot(abs(np.array(raw_results[E][i]["E_value"])-float(E))/float(E) *100)
        plt.ylim([-1,100])
        plt.title(f"Porcenraje de Error Relativo (E_real: {E}) ")
        print(f"Real:{E}, init: {i} Predicho: {raw_results[E][i]['E_value'][-1]:.3f}, RAE: {abs(np.array(raw_results[E][i]['E_value'][-1])-float(E))/float(E) *100:.2f} %")
plt.tight_layout()
```
